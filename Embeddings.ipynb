```
#define metric
def roc_auc(preds, y):
    #round predictions to the closest integer
    score = roc_auc_score(y, np.round(preds,2)) 
    return score
```

```
def create_embedding_matrix(word_index,embedding_dict,dimension):
  embedding_matrix=np.zeros((len(word_index),dimension))
  k=0
  for word,index in word_index.items():
    if word in embedding_dict:
      embedding_matrix[index]=embedding_dict[word]
  return embedding_matrix
```

```
def dividing_data(data, n):
     '''
     Функция для разделения данныех в соответствии с соотношением toxic & clean
     data - таблица на вход и выход
     '''
     x = int(len(data[data.toxic==0])*n/len(data))
     y = int(len(data[data.toxic==1])*n/len(data))
     x = data[data.toxic==0].sample(x,random_state=42)
     y = data[data.toxic==1].sample(y,random_state=42)
     data = pd.concat([x,y], ignore_index=True)
     return data
```

```
def text_cleaning(text, lang):
    '''
    Функция для чистки текста. 
    text - текст, над которым производится чистка
    '''
    def strip_html_tags(text):
        '''
        Функция для удаления html tags
        '''
        soup = BeautifulSoup(text, "html.parser")
        stripped_text = soup.get_text(separator=" ")
        return stripped_text

    def remove_URL(text):
        '''
        Функция для удаления url.
        '''
        url = re.compile(r'http(s*)?://\S+|www\.\S+')
        return url.sub(r'',text)

    def remove_mentions(text):
        '''
        Функция для удаления url.
        '''
        return re.sub(r'@[\S]+\b', '', text)

    def remove_emoji(text):
        '''
        Функция для удаления эмоджис и эмотиконов в тексте.
        '''
        emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', text)

    def remove_ip(text):
        '''
        Функция для удаления ip адресов
        '''
        text = re.sub("([0-9]{1,3}[\.]){3}[0-9]{1,3}", "", text)
        return text

    def remove_accented_chars(text):
        '''
        Функция для удаления accented characters из текста, e.g. café.
        '''
        text = unidecode.unidecode(text)
        return text

    def decontracted(text):
        '''
        Функция для работы с сокращениями
        '''
        # specific
        text = re.sub(r"won\'t", "will not", text)
        text = re.sub(r"can\'t", "can not", text)
        # general
        text = re.sub(r"n\'t", " not", text)
        text = re.sub(r"\'re", " are", text)
        text = re.sub(r"\'s", " is", text)
        text = re.sub(r"\'d", " would", text)
        text = re.sub(r"\'ll", " will", text)
        text = re.sub(r"\'t", " not", text)
        text = re.sub(r"\'ve", " have", text)
        text = re.sub(r"\'m", " am", text)
        return text

    def correct_spelling(text):
         '''
         Функция для spellcheking
         text - строка текста
         '''
         if spell == 'unknown':
             pass
         else:
             corrected_text = []
             misspelled_words = spell.unknown(text.split())
             for word in text.split():
                 if word in misspelled_words:
                     corrected_text.append(spell.correction(word))
                 else:
                     corrected_text.append(word)

             text = " ".join(corrected_text)
         return text

    def remove_stopwords(text):
         '''
         Функция для удаления стоп-слов
         text - строка текста
         '''
         text_new = []    
         tokens = word_tokenize(text)
         tokens = [token for token in tokens if ((len(token) >= 2) and (token not in stop_words))]
         tokens = ' '.join(tokens)
         text_new.append(tokens)
         return text_new[0]

    def lemmatization(text):
         '''
         Функция для лемматизации
         text - строка текста
         '''
         if nlp == 'unknown':
             pass
         else:
             text = nlp(text)
             st = []
             for element in text:
                st.append(element.lemma_)
             text = ' '.join(st)
         return text

    filters = [ 
             lambda x: x.lower(),
             strip_html_tags,
             remove_URL, 
             remove_emoji,
             remove_ip,
             remove_mentions,
             decontracted, #сокращения на английском
             gsp.strip_punctuation, # уберем пунктуацию. Большинство текстов небольшой длины. В данном случае контекстом внутри предложений можно пренебречь.
             gsp.strip_numeric, # уберем цифры
             gsp.strip_multiple_whitespaces, # уберем множественные пробелы
             lambda x: x.strip(), # уберем начальные и конечные пробелы
             correct_spelling,
             remove_stopwords,
             lemmatization,
             ]

    # выберем инструментарий в зависимости от языка
    if lang == 'en':
          spell = SpellChecker(language='en', distance=1)
          stop_words = stop_words_en
          nlp =  nlp_en
    elif lang == 'fr':
          spell = SpellChecker(language='fr', distance=1)
          stop_words = stop_words_fr
          nlp =  nlp_fr
    elif lang == 'es':
          spell = SpellChecker(language='es', distance=1)
          stop_words = stop_words_es
          nlp =  nlp_es
    elif lang == 'it':
          spell = 'unknown'
          stop_words = stop_words_it
          nlp =  nlp_it
    elif lang == 'pt':
          spell = SpellChecker(language='pt', distance=1)
          stop_words = stop_words_pt
          nlp =  nlp_pt
    elif lang == 'ru':
          spell = 'unknown' 
          # spell check на русском делать не будет - слишком большая вариативность языка
          stop_words = stop_words_ru
          nlp =  nlp_ru
    else:
          spell = 'unknown'
          stop_words = stop_words_tr
          nlp =  'unknown'

    text = utils.to_unicode(text)
    text = sent_tokenize(text)

    for f in filters:
          text = [f(row) for row in text]
                
    return text
```

##word2vec

Не нашел предобученных эмбеддингов для нескольких языков. Рассматривал 2 варианта: стакать матрицы эмбеддингов и сделать ансамбль моделей. Решил сделать ансамбль.

```
class Classifier(nn.ModuleList):
  def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim, n_layers, 
              bidirectional, dropout):
    super(Classifier, self).__init__()
    self.dropout = nn.Dropout(0.5)
    self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=1)
    self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers=n_layers,
                        bidirectional=bidirectional, dropout=dropout, 
                        batch_first=True)
    self.fc = nn.Linear(in_features=hidden_dim*2, out_features=output_dim)
    self.act = nn.Sigmoid()
  
  def forward(self, text, text_lengths):
    
    #text = [batch size,sent_length]
    embedded = self.embedding(text)
    #embedded = [batch size, sent_len, emb dim]
  
    #packed sequence
    packed_embedded = nn.utils.rnn.pack_padded_sequence(embedded, text_lengths.cpu(),
                                                        batch_first=True)
    
    packed_output, (hidden, cell) = self.lstm(packed_embedded)
    #hidden = [batch size, num layers * num directions,hid dim]
    #cell = [batch size, num layers * num directions,hid dim]
    
    #concat the final forward and backward hidden state
    hidden = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim = 1)
            
    #hidden = [batch size, hid dim * num directions]
    dense_outputs=self.fc(hidden)

    #Final activation function
    outputs=self.act(dense_outputs)
    
    return outputs
```

```
def train(model, iterator, optimizer, criterion):
    
    #initialize every epoch 
    epoch_loss = 0
    epoch_acc = 0
    
    #set the model in training phase
    model.train()  
    y_true = []
    y_preds = []
    for batch in tqdm(iterator):
        
        #resets the gradients after every batch
        optimizer.zero_grad()   
        
        #retrieve text and no. of words
        text, text_lengths = batch.correct_text   
        
        #convert to 1D tensor
        predictions = model(text, text_lengths).squeeze()  
        
        #compute the loss
        loss = criterion(predictions, batch.toxic)        
        y_true += batch.toxic.tolist()
        y_preds += predictions.tolist()
        
        #backpropage the loss and compute the gradients
        loss.backward()       
        
        #update the weights
        optimizer.step()      
        
        #loss and accuracy
        epoch_loss += loss.item()  

    epoch_acc = roc_auc(y_preds, y_true)   
        
    return epoch_loss / len(iterator), epoch_acc
```

```
def evaluate(model, iterator, criterion):
    
    #initialize every epoch
    epoch_loss = 0
    epoch_acc = 0

    #deactivating dropout layers
    model.eval()
    y_true = []
    y_preds = []
    #deactivates autograd
    with torch.no_grad():

      for batch in iterator:

        #retrieve text and no. of words
        text, text_lengths = batch.correct_text
        
        #convert to 1d tensor
        predictions = model(text, text_lengths).squeeze()
        
        #compute loss and accuracy
        loss = criterion(predictions, batch.toxic)
        y_true += batch.toxic.tolist()
        y_preds += predictions.tolist()
        #keep track of loss and accuracy
        epoch_loss += loss.item()

    epoch_acc = roc_auc(y_preds, y_true) 
        
    return epoch_loss / len(iterator), epoch_acc
```

```
def model_training_for_lang(lang,embedding_path,training_data,val_data=None, n_epoch=5,req_grad=False):
  train_data = training_data[training_data.lang==lang]
  train_data = train_data[train_data.correct_text!='']
  if lang not in val_data.lang.value_counts():
    train_data, val_data = train_test_split(train_data, stratify = train_data.toxic)
  else:
    val_data = val_data[val_data.lang==lang]
  train_data.to_csv("train_{}.csv".format(lang), index=False)
  val_data.to_csv("val_{}.csv".format(lang), index=False)

  LABEL = legacy.data.LabelField(dtype=torch.float, sequential=True,tokenize=None)
  TEXT = legacy.data.Field(dtype = torch.long,batch_first=True,include_lengths=True)
  fields = [(None, None), ('toxic', LABEL), (None, None), ('correct_text',TEXT)]

  train_data,val_data = legacy.data.TabularDataset.splits(path='./', train="train_{}.csv".format(lang), validation="val_{}.csv".format(lang), format='csv', fields=fields,skip_header=True)
  if lang=='ru':
    TEXT.build_vocab(train_data,min_freq=3)  
  else:
    logger = logging.getLogger('my-logger')
    logger.propagate = False
    vectors = vocab.Vectors(embedding_path)
    TEXT.build_vocab(train_data,min_freq=3,vectors = vectors)
    logger.propagate = True

  LABEL.build_vocab(train_data)
  device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')  

  #set batch size
  BATCH_SIZE = 128

  #Load an iterator
  train_iterator, valid_iterator = legacy.data.BucketIterator.splits(
      (train_data, val_data), 
      batch_size = BATCH_SIZE,
      sort_key = lambda x: len(x.correct_text),
      sort_within_batch=True,
      device = device)
  
  #define hyperparameters
  size_of_vocab = len(TEXT.vocab)
  embedding_dim = 300
  num_hidden_nodes = 32
  num_output_nodes = 1
  num_layers = 2
  bidirection = True
  dropout = 0.2

  #instantiate the model
  model = Classifier(size_of_vocab, embedding_dim, num_hidden_nodes, 
                    num_output_nodes, num_layers, 
                    bidirectional = True, dropout = dropout)
  if lang=='ru':
    dct = TEXT.vocab.stoi
    pretrained_embeddings = create_embedding_matrix(dct, navec, 300)
  else:
    pretrained_embeddings = TEXT.vocab.vectors
    model.embedding.weight.data.copy_(pretrained_embeddings)
  
  model.embedding.weight.requires_grad=req_grad
  #define optimizer and loss
  optimizer = optim.Adam(model.parameters())
  criterion = nn.BCELoss()
      
  #push to cuda if available
  model = model.to(device)
  criterion = criterion.to(device)
  N_EPOCHS = n_epoch
  best_valid_loss = float('inf')

  for epoch in range(N_EPOCHS):
      
      #train the model
      train_loss, train_acc = train(model, train_iterator, optimizer, criterion)
      
      #evaluate the model
      valid_loss, valid_acc = evaluate(model, valid_iterator, criterion)
      
      #save the best model
      if valid_loss < best_valid_loss:
          best_valid_loss = valid_loss
          torch.save(model.state_dict(), 'saved_weights_{}.pt'.format(lang))
      
      print(f'\tTrain Loss: {train_loss:.3f} | Train Acc: {train_acc*100:.2f}%')
      print(f'\t Val. Loss: {valid_loss:.3f} |  Val. Acc: {valid_acc*100:.2f}%')
  
  return TEXT,LABEL,model
```

```
model.eval();

def predict(model, sentence, TEXT):
    tokenized = sentence.split()  #tokenize the sentence 
    indexed = [TEXT.vocab.stoi[t] for t in tokenized]          #convert to integer sequence
    length = [len(indexed)]                                    #compute no. of words
    tensor = torch.LongTensor(indexed).to(device)              #convert to tensor
    tensor = tensor.unsqueeze(1).T                             #reshape in form of batch,no. of words
    length_tensor = torch.LongTensor(length)                   #convert to tensor
    prediction = model(tensor, length_tensor)                  #prediction 
    return prediction.item()  
```

```
# Save and Load Functions

def save_checkpoint(save_path, model, valid_loss):

    if save_path == None:
        return
    
    state_dict = {'model_state_dict': model.state_dict(),
                  'valid_loss': valid_loss}
    
    torch.save(state_dict, save_path)
    print(f'Model saved to ==> {save_path}')

def load_checkpoint(load_path, model):
    
    if load_path==None:
        return
    
    state_dict = torch.load(load_path, map_location=device)
    print(f'Model loaded from <== {load_path}')
    
    model.load_state_dict(state_dict['model_state_dict'])
    return state_dict['valid_loss']


def save_metrics(save_path, train_loss_list, valid_loss_list, global_steps_list):

    if save_path == None:
        return
    
    state_dict = {'train_loss_list': train_loss_list,
                  'valid_loss_list': valid_loss_list,
                  'global_steps_list': global_steps_list}
    
    torch.save(state_dict, save_path)
    print(f'Model saved to ==> {save_path}')


def load_metrics(load_path):

    if load_path==None:
        return
    
    state_dict = torch.load(load_path, map_location=device)
    print(f'Model loaded from <== {load_path}')
    
    return state_dict['train_loss_list'], state_dict['valid_loss_list'], state_dict['global_steps_list']
```

```
class BERT(nn.Module):

    def __init__(self):
        super(BERT, self).__init__()

        options_name = 'distilbert-base-multilingual-cased'
        self.encoder = DistilBertForSequenceClassification.from_pretrained(options_name)

    def forward(self, text, label=None, test=False):
      if test:
        text_fea = self.encoder(text)
        return text_fea
      else:
        loss, text_fea = self.encoder(text, labels=label)[:2]
        return loss, text_fea
```

```
def train(model,
          optimizer,
          criterion = nn.BCELoss(),
          train_loader = train_iter,
          valid_loader = valid_iter,
          num_epochs = 5,
          eval_every = len(train_iter) // 2,
          file_path = './',
          best_valid_loss = float("Inf")):
    
    # initialize running values
    running_loss = 0.0
    valid_running_loss = 0.0
    global_step = 0
    train_loss_list = []
    valid_loss_list = []
    global_steps_list = []

    # training loop
    model.train()
    for epoch in range(num_epochs):
      y_pred_train=[]
      y_pred_valid=[]
      y_true_train = []
      y_true_valid = []
      for text, labels in tqdm(train_loader):
          labels = labels.type(torch.LongTensor)           
          labels = labels.to(device)
          text = text.type(torch.LongTensor)  
          text = text.to(device)
          output = model(text, labels)
          loss, logits = output
          y_pred_train.extend(torch.softmax(logits,1)[:,1].cpu().detach().numpy())
          y_true_train.extend(labels.cpu().detach().numpy())

          optimizer.zero_grad()
          loss.backward()
          optimizer.step()

          # update running values
          running_loss += loss.item()
          global_step += 1

          # evaluation step
          if global_step % eval_every == 0:
              model.eval()
              with torch.no_grad():                    

                  # validation loop
                  for text, labels in valid_loader:
                      labels = labels.type(torch.LongTensor)           
                      labels = labels.to(device)
                      text = text.type(torch.LongTensor)  
                      text = text.to(device)
                      output = model(text, labels)
                      loss, logits = output
                      y_pred_valid.extend(torch.softmax(logits,1)[:,1].cpu().detach().numpy())
                      y_true_valid.extend(labels.cpu().detach().numpy())
                      valid_running_loss += loss.item()

              # evaluation
              average_train_loss = running_loss / eval_every
              average_valid_loss = valid_running_loss / len(valid_loader)
              train_loss_list.append(average_train_loss)
              valid_loss_list.append(average_valid_loss)
              global_steps_list.append(global_step)
              auc_train=roc_auc(y_pred_train,y_true_train)
              auc_valid=roc_auc(y_pred_valid,y_true_valid)
              # resetting running values
              running_loss = 0.0                
              valid_running_loss = 0.0
              model.train()

              # print progress
              print('Epoch [{}/{}], Step [{}/{}], Train Loss: {:.4f}, Valid Loss: {:.4f}, Train AUC: {:.4f}, Valid AUC: {:.4f}'
                    .format(epoch+1, num_epochs, global_step, num_epochs*len(train_loader),
                            average_train_loss, average_valid_loss, auc_train, auc_valid))
              
              # checkpoint
              if best_valid_loss > average_valid_loss:
                  best_valid_loss = average_valid_loss
                  save_checkpoint(file_path + '/' + 'model_dist_1.pt', model, best_valid_loss)
                  save_metrics(file_path + '/' + 'metrics.pt', train_loss_list, valid_loss_list, global_steps_list)
  
    save_metrics(file_path + '/' + 'metrics.pt', train_loss_list, valid_loss_list, global_steps_list)
    print('Finished Training!')
```

```
class ELMO(nn.ModuleList):
  def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim,
              bidirectional, dropout):
    super(ELMO, self).__init__()
    # self.n_lstm_layers = n_layers
    self.hidden_dim = hidden_dim
    if bidirectional:
      self.bilstm = 2
    else:
      self.bilstm = 1
    self.dropout = nn.Dropout(0.5)
    self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=1)
    self.lstm_1 = nn.LSTM(embedding_dim, hidden_dim, 
                        bidirectional=bidirectional, dropout=dropout, 
                        batch_first=True)
    self.lstm_2 = nn.LSTM(self.bilstm*hidden_dim, hidden_dim, 
                        bidirectional=bidirectional, dropout=dropout, 
                        batch_first=True)
    self.fc = nn.Linear(in_features=hidden_dim*2, out_features=output_dim)
    self.act = nn.Sigmoid()
  
  def forward(self, text, text_lengths):
    
    #text = [batch size,sent_length]
    embedded = self.embedding(text)
    #embedded = [batch size, sent_len, emb dim]
    # print(embedded.size())
  
    #packed sequence
    # packed_embedded = nn.utils.rnn.pack_padded_sequence(embedded, text_lengths.cpu(),
    #                                                 batch_first=True)

    # print(packed_embedded.data.size())
    
    packed_output_1, (hidden_1, cell_1) = self.lstm_1(embedded)
    packed_output_2, (hidden_2, cell_2) = self.lstm_1(packed_output_1,(hidden_1, cell_1))

    # packed_output_1 = torch.nn.functional.softmax(packed_output_1.data,dim=1)
    # packed_output_1 = packed_output_1.reshape(embedded.size())
    # packed_output_1 = torch.nn.functional.softmax(packed_output_1.data.reshape(embedded.size()),dim=1)
    # packed_output_2 = torch.nn.functional.softmax(packed_output_2.data,dim=1)
    # packed_output_2 = packed_output_2.reshape(embedded.size())
    # packed_output_2 = torch.nn.functional.softmax(packed_output_2.data.reshape(embedded.size()),dim=1)
    # embedded = torch.nn.functional.softmax(embedded,dim=1)

    new_emb = packed_output_1 + packed_output_2 + embedded
    new_emb = new_emb.mean(dim=1)
    # packed_output, (hidden, cell) = self.lstm_2(packed_output)
    # # hidden.view(batch, self.n_lstm_layers, self.bilstm, self.hidden_dim)
    # #hidden = [batch size, num layers * num directions,hid dim]
    # #cell = [batch size, num layers * num directions,hid dim]
    
    # #concat the final forward and backward hidden state
    # hidden = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim = 1)
    # hid_l_2 = torch.cat((res[0,:,:],res[1,:,:]),dim=1)
    # hid_l_2 = torch.cat((res[-2,:,:],res[-1,:,:]),dim=1)        
    # #hidden = [batch size, hid dim * num directions]
    dense_outputs=self.fc(new_emb)

    # #Final activation function
    outputs=self.act(dense_outputs)
    
    return outputs
```

```
def train(model, iterator, optimizer, criterion):
    
    #initialize every epoch 
    epoch_loss = 0
    epoch_acc = 0
    
    #set the model in training phase
    model.train()  
    y_true = []
    y_preds = []
    for batch in tqdm(iterator):
        
        #resets the gradients after every batch
        optimizer.zero_grad()   
        
        #retrieve text and no. of words
        text, text_lengths = batch.correct_text   
        
        #convert to 1D tensor
        predictions = model(text, text_lengths).squeeze()  
        
        #compute the loss
        loss = criterion(predictions, batch.toxic)        
        if (len(batch.toxic)==1):
          loss = criterion(predictions, batch.toxic[0])
          y_true.append(batch.toxic[0].cpu().numpy().tolist())
          y_preds.append(predictions.tolist())
        else:
          loss = criterion(predictions, batch.toxic)
          y_true += batch.toxic.tolist()
          y_preds += predictions.tolist() 
        
        #backpropage the loss and compute the gradients
        loss.backward()       
        
        #update the weights
        optimizer.step()      
        
        #loss and accuracy
        epoch_loss += loss.item()  

    epoch_acc = roc_auc(y_preds, y_true)   
        
    return epoch_loss / len(iterator), epoch_acc
```

```
def evaluate(model, iterator, criterion):
    
    #initialize every epoch
    epoch_loss = 0
    epoch_acc = 0

    #deactivating dropout layers
    model.eval()
    y_true = []
    y_preds = []
    #deactivates autograd
    with torch.no_grad():

      for batch in iterator:

        #retrieve text and no. of words
        text, text_lengths = batch.correct_text
        
        #convert to 1d tensor
        predictions = model(text, text_lengths).squeeze()
        
        #compute loss and accuracy
        loss = criterion(predictions, batch.toxic)
        y_true += batch.toxic.tolist()
        y_preds += predictions.tolist()
        #keep track of loss and accuracy
        epoch_loss += loss.item()

    epoch_acc = roc_auc(y_preds, y_true) 
        
    return epoch_loss / len(iterator), epoch_acc
```

```
def model_training_for_lang(lang,embedding_path,training_data,val_data=None, n_epoch=5,req_grad=False):

  train_data = training_data[training_data.lang==lang]
  train_data = train_data[train_data.correct_text!='']

  if lang not in val_data.lang.value_counts():
    train_data, val_data = train_test_split(train_data, stratify = train_data.toxic)
  else:
    val_data = val_data[val_data.lang==lang]

  train_data.to_csv("train_{}.csv".format(lang), index=False)
  val_data.to_csv("val_{}.csv".format(lang), index=False)

  LABEL = legacy.data.LabelField(dtype=torch.float, sequential=True,tokenize=None)
  TEXT = legacy.data.Field(dtype = torch.long,batch_first=True,include_lengths=True)
  fields = [(None, None), ('toxic', LABEL), (None, None), ('correct_text',TEXT)]

  train_data,val_data = legacy.data.TabularDataset.splits(path='./', train="train_{}.csv".format(lang), validation="val_{}.csv".format(lang), format='csv', fields=fields,skip_header=True)

  if lang=='ru':
    TEXT.build_vocab(train_data,min_freq=3)  
  else:
    logger = logging.getLogger('my-logger')
    logger.propagate = False
    vectors = vocab.Vectors(embedding_path)
    TEXT.build_vocab(train_data,min_freq=3,vectors = vectors)
    logger.propagate = True

  LABEL.build_vocab(train_data)
  device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')  

  #set batch size
  BATCH_SIZE = 128

  #Load an iterator
  train_iterator, valid_iterator = legacy.data.BucketIterator.splits(
      (train_data, val_data), 
      batch_size = BATCH_SIZE,
      sort_key = lambda x: len(x.correct_text),
      sort_within_batch=True,
      device = device)
  
  #define hyperparameters
  size_of_vocab = len(TEXT.vocab)
  embedding_dim = 300
  num_hidden_nodes = 150
  num_output_nodes = 1
  num_layers = 2
  bidirection = True
  dropout = 0.2

  #instantiate the model
  model = ELMO(size_of_vocab, embedding_dim, num_hidden_nodes, 
                    num_output_nodes, 
                    bidirectional = True, dropout = dropout)
  if lang=='ru':
    dct = TEXT.vocab.stoi
    pretrained_embeddings = create_embedding_matrix(dct, navec, 300)
  else:
    pretrained_embeddings = TEXT.vocab.vectors
    model.embedding.weight.data.copy_(pretrained_embeddings)
  
  model.embedding.weight.requires_grad=req_grad
  #define optimizer and loss
  optimizer = optim.Adam(model.parameters())
  criterion = nn.BCELoss()
      
  #push to cuda if available
  model = model.to(device)
  criterion = criterion.to(device)
  N_EPOCHS = n_epoch
  best_valid_loss = float('inf')

  for epoch in range(N_EPOCHS):
      
      #train the model
      train_loss, train_acc = train(model, train_iterator, optimizer, criterion)
      
      #evaluate the model
      valid_loss, valid_acc = evaluate(model, valid_iterator, criterion)
      
      #save the best model
      if valid_loss < best_valid_loss:
          best_valid_loss = valid_loss
          torch.save(model.state_dict(), 'saved_weights_{}.pt'.format(lang))
      
      print(f'\tTrain Loss: {train_loss:.3f} | Train Acc: {train_acc*100:.2f}%')
      print(f'\t Val. Loss: {valid_loss:.3f} |  Val. Acc: {valid_acc*100:.2f}%')
  
  return TEXT,LABEL,model
```
